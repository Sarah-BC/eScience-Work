## Introduction to plastic pollution

As of 2011, the global demand for plastics was estimated at 245 million tonnes.  At the same time, roughly 1/3 of the plastic being produced was destined to become single-use, disposable products of the sort often encountered on beaches and roadsides around the world (Andrady, 2011).  It is no wonder that much of this plastic, as well as plastic from a variety of other sources, finds its way into our oceans.  The task of quantifying (counting or measuring) that amount, and understanding the affects this plastic pollution has on the marine environment is a daunting one.  With this notebook, we intend to gather as much of the current research on marine plastic pollution as we can find and to communicate it in a way that is accessible to people of all backgrounds.  

But what do we mean when we talk about plastic?  And what is it that makes plastic pollution in our oceans so concerning for researchers and community scientists across the world? While plastics can now come in compostible and biodegradable forms, beginning with their discovery in 1907, plastics have been designed to resist breaking down either mechanically (grinding, cracking, or tearing into smaller chunks and fibers) or chemically (breaking the bonds between molecules and atoms within the plastic). This ability to resist breaking down in the environment is a key characteristic of plastics and is a large part of what makes them such pervasive debris. 

When plastics get into the ocean they can stick around in a variety of ecosystems for many, many years.  While the sunlight, oxygen, and wave action at the surface of the ocean can begin the slow degredation process, once the plastics move into deeper water, this process slows to a near halt.  And finding out how quickly plastics degrade in the ocean is as complex an answer as it is a simple question.  Plastics come in a wide range of forms and functions, all of which affect the speed of physical and chemical breakdown, and each region of the oceans has unique characteristics that further affect the rates of decay.  This entire process can be aided by, and be harming the marine life in each of these environments.

While it is easy to assume that plastic pollution, like all pollution to different extents, is harmful for the environment it enters, qualifying (describing) and quantifying that harm is yet another difficult task.  The oceans contain the smallest form of life in microscopic achaea and bacteria, but also the largest lifeforms, whales and collosal squid.  Every single form of life in the ocean effects and is effected by plastic differently.  For Green Sea Turtles and sperm whales, plastic bags look so similar to jelly fish that they can be eaten by mistake and choke the creatures or fill their stomachs to the point of not being able to intake proper nutrients. On the opposite end of the spectrum, bacterial colonies can thrive on plastic debris that acts as an island for them to anchor to in their passive search for food.  And in the relative middle of the size range, many fish can consume  pre-production microplastic pellets, called nurdles, and pass them through their digestive tracks with no apparent harm.  

The questions surrounding plastic pollution in the sea seem as vast as the ocean's themselves. ### Characteristics of plastic pollution

Not all plastics are created equal.  Polyethyelene (PE), Polypropylene (PP), Polystyrene (PS), Poly(ethylene terephthalate) (PET), and Poly(vinyl chloride) (PVC) are all different forms of plastics used in the creation of common goods and packaging.  Each of these compounds has a unique chemical structure and different physical attributes.  What most synthetic (not naturally occuring) plastics have in common is a resistance to breaking down and an ability to maintain their form and/or funtion even in harsh environments. 

In addition to having different chemical structures, some plastics are treated with additional chemicals, often Persistant Organic Pollutants (POPs), in order to achieve new properties such as:
 - Flame retardant
 - Insect repellent
 - Add or reduce scents
 
 
 These POP's have also been found to have some serious effects on the biology of animals and plants that absorb them through digestion, physical contact, or leaching of the POP into waters.
